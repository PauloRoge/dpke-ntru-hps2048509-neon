```python
aZ = Z(a)(2)
bZ = Z(b)(2)
cZ = Z(c)(2)
cl = cZ % 2**64
ch = cZ // 2**64
print(f"{{ 0x{aZ:016x}, 0x{bZ:016x}, 0x{cl:016x}, 0x{ch:016x} }},")
```

```python
mask = (1 << 64) - 1

a = Z2.random_element(degree=127)
b = Z2.random_element(degree=127)

c = a * b

aZ = Z(a)(2)
bZ = Z(b)(2)
cZ = Z(c)(2)

# quebra entradas 128 em 2x64
a0 =  aZ        & mask
a1 = (aZ >> 64) & mask

b0 =  bZ        & mask
b1 = (bZ >> 64) & mask

# quebra saÃ­da 256 em 4x64
r0 =  cZ         & mask
r1 = (cZ >>  64) & mask
r2 = (cZ >> 128) & mask
r3 = (cZ >> 192) & mask

print("a(x) =", a)
print("b(x) =", b)
print("c(x) =", c)

# linha pronta para um struct do teste (ajuste o nome/campos conforme seu testcase_128)
print(
    f"{{UINT64_C(0x{a0:016x}), UINT64_C(0x{a1:016x}), "
    f"UINT64_C(0x{b0:016x}), UINT64_C(0x{b1:016x}),\n"
    f"     UINT64_C(0x{r0:016x}), UINT64_C(0x{r1:016x}), "
    f"UINT64_C(0x{r2:016x}), UINT64_C(0x{r3:016x})}},"
)
```
